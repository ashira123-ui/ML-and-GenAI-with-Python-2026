```python
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score,confusion_matrix

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Dataset 2.csv")
print(df.head())
```

```python
print("Number of rows and columns in dataset:",df.shape)
```

```python
print("Column names:", df.columns)
```

```python
print("Numerical and categorical values:\n",df.dtypes)
```

```python
print("Average age of users:", df["Age"].mean())
```

```python
print("Average watch hours per week:", df["WatchHoursPerWeek"].mean())
```

```python
print("Average monthly spending of users:", round(df["MonthlySpend"].mean(),2))
```

```python
print("Number of users in each subscription category:\n",df.SubscriptionType.value_counts())
```

```python
print("Percentage of users who renewed their subscription:",round((df['SubscriptionRenewed']=='Yes').mean()*100,2))
```

```python
le = LabelEncoder()
df["Gender"]=le.fit_transform(df['Gender'])
df["SubscriptionType"] = le.fit_transform(df['SubscriptionType'])
df["FavoriteGenre"] = le.fit_transform(df['FavoriteGenre'])
```

```python
x = df.drop(["UserID","SubscriptionRenewed"],axis=1)
y = df["SubscriptionRenewed"]

x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)

dt=DecisionTreeClassifier(random_state=42)
dt.fit(x_train,y_train)
pred = dt.predict(x_test)

print("Decision tree accuracy:", accuracy_score(y_test,pred))
print("Confusion matrix:\n", confusion_matrix(y_test,pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
knn_pred= knn.predict(x_test)
print("KNN accuracy:",accuracy_score(y_test,knn_pred))
```

```python
x_reg=df.drop(["UserID","SubscriptionRenewed"],axis=1)
y_reg=df["MonthlySpend"]

x_train_reg,x_test_reg,y_train_reg,y_test_reg=train_test_split(
    x_reg,y_reg,test_size=0.2,random_state=42
  )

lr=LinearRegression()
lr.fit(x_train_reg,y_train_reg)
input_data=pd.DataFrame([[25,0,0,15,3,1,12,220]],columns=x_reg.columns)
amount = lr.predict((input_data))
print("Predicting Monthly spending:", amount[0])
```
